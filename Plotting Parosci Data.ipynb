```python
%matplotlib inline

import os
import numpy as np
import matplotlib.pyplot as plt

os.chdir('/home/mschwehr/multibeam_data/20161118RachelCarson/d20161118/')
```

```python
time_p = []
press_p = []
f = open('20161118_162929_parosci.txt','r')
lines=f.readlines()
for x in lines:
    time_p.append(x.split(' ')[0])
    press_p.append(x.split(' ')[1])
```

```python
plt.plot(time_p,press_p)
```

```python
time_d = []
sensor_d = []
f2 = open('20161118_Ventana_INS_sensordepth_Mc.txt','r')
lines2=f2.readlines()
for x in lines2:
    time_d.append(x.split('\t')[0])
    sensor_d.append(x.split('\t')[1])
```

```python
plt.plot(time_d,sensor_d)
```

```python
plt.plot(time_p,press_p)
plt.plot(time_d,sensor_d)
plt.legend(['parosci','sensor depth'],loc=4)
```

```python
tdes = []
for root, dirs, files in os.walk(r'/home/mschwehr/multibeam_data/20161118RachelCarson/d20161118/'):
    for file in files:
        if file.endswith('.tde'):
            tdes.append(file)
            
tdes.sort()
```

```python
time_t = []
tide_t = []
for tide in tdes:
    f3 = open(tide,'r')
    lines3=f3.readlines()
    for x in lines3:
        if x.startswith('#'):
            continue
        time_t.append(x.split('    ')[0])
        tide_t.append(x.split('    ')[1].strip('\n'))
    f3.close()
```

```python
plt.plot(time_t,tide_t)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot(time_p,press_p)
ax.plot(time_d,sensor_d)
ax.invert_yaxis()

ax.set_ylabel("Depth (m)")
ax.set_xlabel("Time (s)")
ax2 = ax.twinx()
ax2.plot(time_t,tide_t,'r-')
ax.legend(['parosci','sensor depth'],loc=2)
ax2.legend(['tide'],loc=6)

ax.spines['right'].set_color('red')
ax2.yaxis.label.set_color('red')
ax2.tick_params(axis='y', colors='red')
ax2.set_ylabel("Tidal Height (m)")
```

```python
time_pitch = []
sensor_pitch = []
f4 = open('20161118_Ventana_INS_pitch_MP.txt','r')
lines4=f4.readlines()
for x in lines4:
    time_pitch.append(x.split('\t')[0])
    sensor_pitch.append(x.split('\t')[1])
```
